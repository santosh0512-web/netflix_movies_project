```python
#Datetime data type: Converting an object type into datatype objects with the to_datetime function then we have a datatime object, can extract various part of data such as year, month and day
movies.duplicated().sum()
```

```python
movies['director'].fillna('Unknown', inplace=True)
movies['cast'].fillna('Unknown', inplace=True)
movies['country'].fillna('Unknown', inplace=True)
movies['rating'].fillna('Not Rated', inplace=True)
```

```python
movies.isnull().sum()
```

```python
#Datetime data type: Converting an object type into datatype objects with the to_datetime function then we have a datatime object, can extract various part of data such as year, month and day
movies['date_added'].dtype
```

```python
movies['date_added'] = movies['date_added'].astype(str).str.strip()
movies['date_added'] = pd.to_datetime(movies['date_added'], errors='coerce')
movies['month_added'] = movies['date_added'].dt.month_name()
month_counts = movies['month_added'].value_counts().reset_index()
month_counts.columns = ['Month', 'Count']
print(month_counts)
```

```python
#Show the all Movies & TV Shows released by month
plt.figure(figsize=(12,5))
plt.bar(month_counts['Month'], month_counts['Count'])
plt.xlabel("Months")
plt.ylabel("Number of Shows")
plt.title("Months vs Count of Netflix Content")
plt.xticks(rotation=45)
plt.show()
```

```python
#Count the all types of unique rating & which rating are with most number
rating_counts = movies['rating'].value_counts()
rating_counts.columns = ['Rating']

print(rating_counts)
```

```python
#Salman, Shah Rukh and Akshay Kumar all movie
salman = movies[movies['cast'].str.contains('Salman Khan', na=False)]
shahrukh = movies[movies['cast'].str.contains('Shah Rukh Khan', na=False)]
akshay = movies[movies['cast'].str.contains('Akshay Kumar', na=False)]

salman[['title','release_year']]
```

```python
#find out the Movies & Series have Maximum time length
movies[['length', 'unit']] = movies['duration'].str.split(' ', expand=True)
movies['length'] = pd.to_numeric(movies['length'], errors='coerce')
#longest movies
movies[movies['unit']=='min'].sort_values('length', ascending=False).head()
```

```python
#most seasons
movies[movies['unit']=='Season'].sort_values('length', ascending=False).head()
```

```python
# Year-on-Year show added on Netflix by its type
movies['year_added'] = movies['date_added'].dt.year
year_type = movies.groupby(['year_added', 'type']).size().reset_index(name='count')
sns.lineplot(x='year_added', y='count',data=year_type, hue='type')
plt.title('Yearly Content Added by Type')
plt.show()
```

```python
#Actors and actresses who have given the most Number of Movies
actors = movies['cast'].str.split(', ', expand=True).stack()
actors = actors.value_counts().head(10)

print(actors)
```

```python
plt.figure(figsize=(14, 4))
sns.barplot(x=actors.index, y=actors.values)
plt.xticks(rotation=45)
plt.title('Count of Actors')
plt.show()
```
